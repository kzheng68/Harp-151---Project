```python
import requests
from bs4 import BeautifulSoup

page = 1

while page != 20:
    source = requests.get(f"https://www.collegexpress.com/college/search/?sortedby=sort_name&keywords=&state=NY&enrollment_min=&enrollment_max=&page={page}").text
    books = BeautifulSoup(source, "lxml")

    for i in books.find_all("div", class_="mb-20"):

        university_name_true = i.find("div", class_="name align-l")
        if university_name_true:
            university_name = university_name_true.text.strip()
        else:
            university_name = None

        location_true = i.find("div", "note align-l mb-10")
        if location_true:
            location = location_true.text.strip()
        else:
            location = None

        tuition_true = i.find("div", class_="info")
        if tuition_true:
            tuition = tuition_true.text.strip()
        else:
            tuition = None

        #to help me specify what tag (the text =), i used this for referencehttps://stackoverflow.com/questions/5999747/beautifulsoup-nextsibling
        acceptance_elem = i.find("span", class_="title", string="ACCEPTANCE RATE")
        acceptance_info = acceptance_elem.find_next_sibling("div", class_="info") if acceptance_elem else None
        acceptance = acceptance_info.text.strip() if acceptance_info else None


        student_teach_ratio_true = i.find("span", class_="title", string="STUDENT-FACULTY RATIO")
        if student_teach_ratio_true:
            student_teach_ratio_info = student_teach_ratio_true.find_next_sibling("div", class_="info")
            if student_teach_ratio_info:
                student_teach_ratio = student_teach_ratio_info.text.strip()

            else:
              student_teach_ratio = None

        else:
            student_teach_ratio_info = None

        desc_true = i.find("div", class_="clear-b")
        if desc_true:
            desc = desc_true.text.strip()
        else:
            desc = None


        print("\n", university_name, "|", location, "|", tuition, "|", acceptance, "|", student_teach_ratio, "|", desc)

    page += 1
```

```python
import requests
from bs4 import BeautifulSoup
import csv

page = 1

with open('colleges.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Name', 'Website', 'Location', 'Tuition', 'Acceptance Rate', 'Student-Faculty Ratio', 'Description']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()

    while page != 20:
        source = requests.get(f"https://www.collegexpress.com/college/search/?sortedby=sort_name&keywords=&state=NY&enrollment_min=&enrollment_max=&page={page}").text
        books = BeautifulSoup(source, "lxml")

        for i in books.find_all("div", class_="mb-20"):

            university_name_true = i.find("div", class_="name align-l")
            university_name = university_name_true.text.strip() if university_name_true else None

            university_url_true = i.a["href"] if i.a else None
            university_url = university_url_true

            location_true = i.find("div", "note align-l mb-10")
            location = location_true.text.strip() if location_true else None

            tuition_true = i.find("div", class_="info")
            tuition = tuition_true.text.strip() if tuition_true else None

            acceptance_elem = i.find("span", class_="title", string="ACCEPTANCE RATE")
            acceptance_info = acceptance_elem.find_next_sibling("div", class_="info") if acceptance_elem else None
            acceptance = acceptance_info.text.strip() if acceptance_info else None

            student_teach_ratio_true = i.find("span", class_="title", string="STUDENT-FACULTY RATIO")
            student_teach_ratio = student_teach_ratio_true.find_next_sibling("div", class_="info").text.strip() if student_teach_ratio_true else None

            desc_true = i.find("div", class_="clear-b")
            desc = desc_true.text.strip() if desc_true else None

            # Write to CSV 
            writer.writerow({
                'Name': university_name,
                'Website': university_url,
                'Location': location,
                'Tuition': tuition,
                'Acceptance Rate': acceptance,
                'Student-Faculty Ratio': student_teach_ratio,
                'Description': desc
            })

        page += 1
```

```python
df1 = pd.read_csv('/colleges_csv.csv')

# Read the second CSV file
df2 = pd.read_csv('/colleges2.csv')


df_master = df1.merge(df2,
                     on = "Name",
                     how = 'outer')

df_master.columns = df_master.columns.str.replace("_x", "")
df_master
```

```python
x_cols = [col for col in df_master.columns if '_y' in col]
x_cols = [s.strip('_y') for s in x_cols]
x_cols
```

```python
for n in x_cols:
    df_master[n] = df_master[n].combine_first(df_master[n +"_y"])


df_master = df_master[df_master.columns.drop(df_master.filter(regex='_y'))]
df_master
```

```python
df_master.to_csv('project_combined_colleges.csv', index=False) #filename
```

```python
with open('project_combined_colleges.csv', 'r', newline='') as file:
    reader = csv.reader(file)
    header = next(reader)  # Read the header row
    data = sorted(reader, key=lambda row: row[0])  # Sort data based on the first column (University Name)

# Write the sorted data back to the combined colleges CSV file
with open('sorted_combined_colleges.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # Write the header row
    writer.writerows(data)

abetcolleges = pd.read_csv("sorted_combined_colleges.csv")
abetcolleges.fillna('Information Unavailable', inplace = True) ## Fills N/A
abetcolleges.to_csv('sorted_combined_colleges.csv', index=False) #filename
abetcolleges
```
