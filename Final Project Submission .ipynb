```python
import csv
from tkinter import Tk, Label, Entry, Button, Listbox, Scrollbar, messagebox, Frame, Text, font
import webbrowser

def search_colleges(event=None, college_index=None):
    search_term = search_entries[college_index].get().lower()
    word_found = False
    suggested_colleges = []
##resource for uploading and usking csv file in tkinter instead of an api key: https://www.w3resource.com/python-exercises/tkinter/python-tkinter-dialogs-and-file-handling-exercise-9.php
    with open("sorted_combined_colleges.csv", "r") as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row
        for row in reader:
            if any(search_term in field.lower() for field in row):
                word_found = True
                suggested_colleges.append(row)  #adds the name as suggestion

    if word_found:
        suggest_listboxes[college_index].delete(0, 'end')
        for college in suggested_colleges:
            suggest_listboxes[college_index].insert('end', college[0])
    else:
        suggest_listboxes[college_index].delete(0, 'end')
        suggest_listboxes[college_index].insert('end', "No suggestions")

def select_suggestion(event=None, college_index=None):
    selected_college = suggest_listboxes[college_index].get('active')
    search_entries[college_index].delete(0, 'end')
    search_entries[college_index].insert('end', selected_college)

def perform_search(college_index=None):
    search_term = search_entries[college_index].get().lower()
    word_found = False

    with open('sorted_combined_colleges.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)  
        for row in reader:
            if any(search_term in field.lower() for field in row):
                word_found = True
#used the concept of the weather GUI to make variables based on row and print it out, so that it'll be cleaner
#and easier to reference for our buttons
                college_name = row[0]
                location = row[1]
                website = row[2]
                largest_program = row[3]
                result_text = f"University Name:\t{college_name}\nLocation:\t{location}\nWebsite:\t{website}\nTuition:\t{largest_program}\n\n"
                result_textbox[college_index].delete(1.0, 'end') #delete the contents inside the textbox if there is any 
                result_textbox[college_index].insert('end', result_text) 
                result_textbox[college_index].update_idletasks()  
#we had trouble with our text not updating before the button used this for help: https://stackoverflow.com/questions/68671470/tkinter-how-to-update-button-text-before-command-function-ends
                show_location_button = Button(result_textbox[college_index], text="Show Location", command=lambda l=location: show_college_location(l))
                show_website_button = Button(result_textbox[college_index], text="Visit Website", command=lambda w=website: open_website(w))
#two buttons follow the same concept as our other buttons, we used the class weather GUI (tkinter day 3)and this as reference: #resource for command=lambda: https://www.geeksforgeeks.org/using-lambda-in-gui-programs-in-python/
#https://stackoverflow.com/questions/69037295/modifying-a-widget-that-was-formerly-embedded-in-a-tkinter-text-widget-produces
                result_textbox[college_index].window_create("end", window=show_location_button)
                result_textbox[college_index].window_create("end", window=show_website_button)
                result_textbox[college_index].insert("end", "\n\n")
                break

    if not word_found:
        result_textbox[college_index].delete(1.0, 'end')
        result_textbox[college_index].insert('end', "No matching college found")
        messagebox.showwarning(title='Not found', message='No matching college found')
        #message box: https://www.geeksforgeeks.org/python-tkinter-messagebox-widget/#

#for the map_url variable we had to use this to help us find a url that would fit into our general location:
#https://stackoverflow.com/questions/2660201/what-parameters-should-i-use-in-a-google-maps-url-to-go-to-a-lat-lon
def show_college_location(location):
    map_url = f"https://www.google.com/maps?q={location}"
    webbrowser.open_new(map_url)
#function to open the hyperlink lines 69 & 72: https://www.geeksforgeeks.org/python-script-to-open-a-web-browser/
def open_website(website):
    webbrowser.open_new(website)
    
#making an instructions using a message box with this for help:
#https://www.geeksforgeeks.org/python-tkinter-messagebox-widget/
def display_instructions():
    messagebox.showinfo("Instructions", "Enter the name of the college you want to learn more about. Double-click on a suggestion to select your desired colleges. Click on the 'Search' button to see the details of the selected colleges. If you're interested, you can click on the 'Show Location' button to get a general location of the area the college is located in. If you click on the ' Visit Website' button you will be directed towards a website that goes into greater detail of the college of your choice.")

#main application window
root = Tk()
root.title("College Search")
root.configure(bg="#f8f9fa")

#our header to display the name of our project
header_label = Label(root, text="College Comparison", font=("Verdana Bold", 20), bg="#f8f9fa", fg= "black")
header_label.grid(row=0, columnspan=3, pady=20)
#instruction button
instructions_button = Button(root, text="Instructions", command=display_instructions)
instructions_button.grid(row=1, column=1, pady=10)

# Create lists to store widgets for each college frame
search_entries = []
suggest_listboxes = []
result_textbox = []

### Create 3 frames for each college comparison with the help of : https://stackoverflow.com/questions/17902619/how-to-create-multiple-frames-in-python-through-for-loop
#and: https://stackoverflow.com/questions/69970356/stack-two-frames-a-frame-with-rows-and-columns-and-a-frame-without
for i in range(3):
    college_frame = Frame(root, padx=10, pady=10, bg="lightblue", borderwidth=2, relief="groove")
    college_frame.grid(row=2, column=i, padx=10, pady=10)

#help csv rows create label https://github.com/hendog993/Python-Youtube/blob/master/CSVtoTKinterLabels.py
    search_label = Label(college_frame, text=f"College {1+i}", bg = "lightblue",fg = "gray11")
    search_label.grid(row=0, column=0, sticky='w')

    search_entry = Entry(college_frame, width=25)
    search_entry.grid(row=1, column=0, sticky='w')
    search_entry.bind('<KeyRelease>', lambda event, index=i: search_colleges(event, index))

    suggest_scrollbar = Scrollbar(college_frame, orient='vertical')
    suggest_scrollbar.grid(row=2, column=1, sticky='ns')

    suggest_listbox = Listbox(college_frame, yscrollcommand=suggest_scrollbar.set, width=30, height=4)
    suggest_listbox.grid(row=2, column=0, sticky='w')
    suggest_listbox.bind('<Double-Button-1>', lambda event, index=i: select_suggestion(event, index))
    #resource for command=lambda: https://www.geeksforgeeks.org/using-lambda-in-gui-programs-in-python/ & tkinter day 3 weather app GUI
    suggest_scrollbar.config(command=suggest_listbox.yview)
       
    #
    search_button = Button(college_frame, text="Search", bg= "white",command=lambda index=i: perform_search(index))
    search_button.grid(row=3, column=0, sticky='w')

    #shows the information for selected colleges
    result_text = Text(college_frame, width=40, height=10)
    result_text.grid(row=4, column=0, sticky='w')

    #add widgets to their lists
    search_entries.append(search_entry)
    suggest_listboxes.append(suggest_listbox)
    result_textbox.append(result_text)

root.mainloop()
 
    
#The Resources that helped the code as a whole:

#Listboxes & textboxes:
#1) https://stackoverflow.com/questions/44958142/tkinter-python-listbox
#2) https://www.geeksforgeeks.org/scrollable-listbox-in-python-tkinter/?ref=next_article
#3) https://python-forum.io/thread-28293.html
```
